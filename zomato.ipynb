# Heatmap of restaurant count on each location

```python
# count restaurants per location
location_count = df['location'].value_counts()

# convert to dataframe
location_df = location_count.reset_index()
location_df.columns = ['location', 'count']

# create heatmap data (pivot)
heatmap_data = location_df.pivot_table(values='count', index='location')

# plot heatmap
plt.figure(figsize=(8,10))
sns.heatmap(heatmap_data, cmap='YlOrRd', annot=False)
plt.title("Heatmap of Restaurant Count by Location")
```

# Heatmap of North Indian restaurantWhat about South Indian cuisines?

```python

# filter cuisines
cuisine_df = df[df['cuisines'].str.contains('North Indian|South Indian', na=False)]

# create pivot table
heatmap_data = pd.crosstab(cuisine_df['location'], cuisine_df['cuisines'].str.extract('(North Indian|South Indian)')[0])

# plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(heatmap_data, cmap='YlOrRd')
plt.title("Heatmap of North Indian vs South Indian Restaurants by Location")
plt.show()
```
